<summary><h4>Zero-shot text classification</h4></summary>

Zero-shot learning is a machine learning paradigm where a model is trained to recognize objects or classes it has never seen during training. In traditional supervised learning, a model is trained on a labeled dataset that includes examples from all the classes it will encounter during inference. However, in zero-shot learning, the model is expected to generalize to new classes or concepts that were not part of its training data.

Here are some key points about zero-shot learning:

Challenge: Zero-shot learning addresses the limitations of traditional supervised learning, which requires exhaustive training data for all possible classes. In many real-world scenarios, it's impractical to collect data for every conceivable class.

Semantic Embeddings: Zero-shot learning often relies on semantic embeddings or continuous vector representations of classes and features. Each class and feature is embedded into a high-dimensional space, where relationships between classes can be captured.

Attributes and Prototypes: Attributes or prototypes are used to describe classes. For example, instead of showing images of different dog breeds during training, a zero-shot learning model might learn attributes like "has fur," "four-legged," "barks," etc., and then combine these attributes to recognize new breeds.

Transductive and Inductive: There are two main approaches to zero-shot learning: transductive and inductive. Transductive zero-shot learning assumes access to some data from unseen classes during inference, while inductive zero-shot learning aims to recognize unseen classes without any additional data.

Applications: Zero-shot learning has applications in image recognition, natural language processing, and recommendation systems. For example, it can be used to recognize new objects in computer vision, translate languages with limited training data, or recommend products in e-commerce for which there's no historical data.

Challenges: Zero-shot learning is challenging because it requires effective knowledge transfer from seen to unseen classes. Handling the semantic gap between textual descriptions and visual features can be difficult. Moreover, selecting appropriate attributes or representations is crucial for success.

Generalization: Successful zero-shot learning models are expected to generalize well to completely novel classes, making them valuable in scenarios where the world is constantly changing and evolving.
</details>

```python
zero_shot_classifier = pipeline("zero-shot-classification")
```

```python
zero_shot_classifier("This laptop has a good battery life",
                     candidate_labels=["education", "tech", "cars"]
                     )
```

#### Mask Filling

```python
unmasker = pipeline('fill-mask')
```

```python
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

```python
unmasker("I am a <mask> engineer", top_k=2)
```

The top_k argument controls how many possibilities you want to be displayed. Note that here the model fills in the special < mask > word, which is often referred to as a mask token. Other mask-filling models might have different mask tokens, so it’s always good to verify the proper mask word when exploring other models. One way to check it is by looking at the mask word used in the widget.

```python
unmasker2 = pipeline("fill-mask", model="bert-base-cased")
```

```python
unmasker2("I am a [MASK] engineer", top_k=2) #[MASK] is the token for this model as opposed to <mask>
```

<summary><h4>Named Entity Recognition</h4></summary>

Named Entity Recognition (NER) is a natural language processing (NLP) technique that focuses on identifying and classifying named entities in text into predefined categories. Named entities are words or phrases that refer to specific objects, such as names of people, organizations, locations, dates, numerical expressions, and more. NER plays a crucial role in extracting structured information from unstructured text data.

Here are the key components and features of Named Entity Recognition:

Entities: NER identifies and categorizes entities into standard types, which typically include:

* Person: Names of individuals.
* Organization: Names of companies, institutions, etc.
* Location: Place names like cities, countries, addresses.
* Date: References to specific dates or date ranges.
* Time: References to specific times or time intervals.
* Percentage: Percentage values.
* Money: Currency amounts.
* Quantity: Numeric quantities or measurements.
* Miscellaneous: Other custom categories as needed.

Applications:

1. Information Extraction: NER is used to extract structured information from unstructured text, which is valuable for various applications like database population, content summarization, and more.
2. Question Answering: In question-answering systems, NER helps identify relevant entities in the question and retrieve corresponding answers.
3. Named Entity Linking: It can be used to link named entities in text to external knowledge bases, providing additional context and information.

Challenges:

1. Ambiguity: Some words can be entities in one context but not in another. For example, "Apple" could refer to the tech company or the fruit.
2. Variability: Names and entities can vary greatly, making it challenging to recognize all possible forms.
3. Out-of-Vocabulary: Handling new or rare entities not seen during training is a challenge.
4. Multilingual NER: Extending NER to multiple languages adds complexity due to language-specific characteristics.

Techniques:

1. Rule-Based: Rule-based NER systems use predefined rules and patterns to recognize entities.
2. Machine Learning: Many modern NER systems use supervised machine learning algorithms, such as Conditional Random Fields (CRF) and deep learning models like Bidirectional LSTMs or Transformers.
3. Pretrained Models: Pretrained language models (e.g., BERT, GPT-3) have achieved state-of-the-art results in NER tasks by fine-tuning on domain-specific data.

Evaluation Metrics:

Common metrics for evaluating NER systems include precision, recall, and F1-score, which measure the system's ability to correctly recognize entities and avoid false positives.
</details>
